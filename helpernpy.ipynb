```python
from copyreg import pickle
import pandas as pd
# from modules.fitness import *
# from modules.helper import *
data = np.load("results_RVE_1_40_D\DB2_GA\initial_simulations2.npy", allow_pickle=True).tolist()
exp_curve = pd.read_csv(f'targets_RVE_1_40_D\DB2.csv')   # <--- Target curve file path.
exp_stress = exp_curve.iloc[:,0] # Getting the experimental stress
exp_strain = exp_curve.iloc[:,1] # Getting the experimental strain
def calculateIndex(stress, strain):
    i = 0
    while strain[i] <= 0.002:
        i += 1
    return stress[i]
expYieldStress = calculateIndex(exp_stress, exp_strain)
print(expYieldStress)
# y = np.array([interpolatedStressFunction(simStress, simStrain, interpolatedStrain) * convertUnit for (simStrain, simStress) in sim.simulations.values()])
# zipParamsStress = list(zip(list(sim.simulations.keys()), y))
sortedClosestYieldStress = list(sorted(data.items(), key=lambda pairs: abs(calculateIndex(pairs[1][1], pairs[1][0]) - expYieldStress)))
sortedClosestYieldStress = dict(sortedClosestYieldStress[:30])
print(len(sortedClosestYieldStress))
print(type(sortedClosestYieldStress))
print(sortedClosestYieldStress)
data = np.save("initial_simulations2.npy", sortedClosestYieldStress)
```
